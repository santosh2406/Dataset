```python
dataFrame = pd.read_csv('C:/Users/e3007251/Desktop/GMF_ETA/GMF_Data_Latest.csv')
```

```python
#Cast data from String to DateTime
dataFrame['Load Start'] = pd.to_datetime(dataFrame['Load Start'], errors='coerce')
dataFrame['ETL Start'] = pd.to_datetime(dataFrame['ETL Start'], errors='coerce')
dataFrame['DeBatch Start'] = pd.to_datetime(dataFrame['DeBatch Start'], errors='coerce')
dataFrame['Reporting Start'] = pd.to_datetime(dataFrame['Reporting Start'], errors='coerce')
dataFrame['Online up Time'] = pd.to_datetime(dataFrame['Online up Time'], errors='coerce')
dataFrame['Load Scheduled'] = pd.to_datetime(dataFrame['Load Scheduled'], errors='coerce')
```

```python
#dataFrame['Duration In Minutes'] = dataFrame['Online up Time'] - dataFrame['Load Start'] 
totalTime = pd.DatetimeIndex(dataFrame['Online up Time'] - dataFrame['Load Start'])
loadTimeGap = pd.DatetimeIndex(dataFrame['Load Start'] - dataFrame['Load Scheduled'])
dataFrame['TotalMinutes'] = totalTime.hour * 60 + totalTime.minute 
dataFrame['Load Start Time Gap'] = loadTimeGap.hour * 60 + loadTimeGap.minute 
dataFrame['SLA Cross'] = np.where(pd.DatetimeIndex(dataFrame['Online up Time']).hour>=6, 'yes', 'no')
```
