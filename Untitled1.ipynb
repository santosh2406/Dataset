```python
import pandas as pd
import numpy as np
from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
from bokeh.io import curdoc
```

```python
gmf_stats = pd.read_csv('C:/Users/e3007251/Desktop/GMF_ETA/BatchJobRunTimes.csv')
```

```python
gmf_stats['Loads Actual Start'] = pd.to_datetime(gmf_stats['Loads Actual Start'], errors='coerce')
gmf_stats['ETL Actual Start'] = pd.to_datetime(gmf_stats['ETL Actual Start'], errors='coerce')
gmf_stats['DMBatch Actual Start'] = pd.to_datetime(gmf_stats['DMBatch Actual Start'], errors='coerce')
gmf_stats['Reporting Actual Start'] = pd.to_datetime(gmf_stats['Reporting Actual Start'], errors='coerce')
gmf_stats['Dialer Outbound'] = pd.to_datetime(gmf_stats['Dialer Outbound'], errors='coerce')
gmf_stats['Triad Outbound US'] = pd.to_datetime(gmf_stats['Triad Outbound US'], errors='coerce')
gmf_stats['Triad Outbound CA'] = pd.to_datetime(gmf_stats['Triad Outbound CA'], errors='coerce')
gmf_stats['Triad Inbound US'] = pd.to_datetime(gmf_stats['Triad Inbound US'], errors='coerce')
gmf_stats['Triad Inbound CA'] = pd.to_datetime(gmf_stats['Triad Inbound CA'], errors='coerce')
gmf_stats['Cycle End Time'] = pd.to_datetime(gmf_stats['Triad Inbound CA'], errors='coerce')
gmf_stats['Total_Load_Time'] = (pd.DatetimeIndex(gmf_stats['Load Time']).hour * 60 + pd.DatetimeIndex(gmf_stats['Load Time']).minute)/60 
gmf_stats['Total_ETL_Time'] = (pd.DatetimeIndex(gmf_stats['ETL Time']).hour * 60 + pd.DatetimeIndex(gmf_stats['ETL Time']).minute)/60 
gmf_stats['Total_DMBatch_Time'] = (pd.DatetimeIndex(gmf_stats['DMBatch Time']).hour * 60 + pd.DatetimeIndex(gmf_stats['DMBatch Time']).minute)/60
gmf_stats['Total_Batch_Cycle'] = (pd.DatetimeIndex(gmf_stats['Total Batch Time']).hour * 60 + pd.DatetimeIndex(gmf_stats['Total Batch Time']).minute)/60 
gmf_stats['SLA_Cross'] = np.where(pd.DatetimeIndex(gmf_stats['Cycle End Time']).hour>=6, 'Y', 'N')
```

```python
#Drop Other date columns
gmf_stats = gmf_stats.drop(columns=['ETL Actual Start','DMBatch Actual Start','Reporting Actual Start','Dialer Outbound','Triad Outbound US','Triad Outbound CA','Triad Inbound US','Triad Inbound CA','Total Batch Time',
'Load Time',
'ETL Time',
'DMBatch Time',
'Reporting Time','Loads Actual Start','Cycle End Time'])
```
